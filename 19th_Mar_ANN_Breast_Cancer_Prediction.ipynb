```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("/content/breast_cancer.csv")
data.head()
```

```python
data.info()
```

```python
## M = Malignant(+ve), B = Benign
data['diagnosis'].value_counts()
```

```python
data.isnull().sum()
```

```python
data = data.drop(["Unnamed: 32","id"],axis=1)
data.head()
```

```python
from sklearn.preprocessing import LabelEncoder
enc = LabelEncoder()
data['diagnosis'] = enc.fit_transform(data['diagnosis'])
data.head()
```

```python
y = data['diagnosis'].values
x = data.drop(['diagnosis'],axis=1).values
```

```python
print(y.shape)
print(x.shape)
```

```python
## Normalize data
```

```python
data.describe()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x = scaler.fit_transform(x)
```

```python
pd.DataFrame(x).describe()
```

```python
## Split the data into train and test
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(x,y,train_size=0.80,random_state=1)
```

```python
print(xtrain.shape)
print(ytrain.shape)
print(xtest.shape)
print(ytest.shape)
```

```python
data['diagnosis'].value_counts()
```

```python
pd.DataFrame(ytrain).value_counts()
```

```python
import tensorflow.keras as tf
```

```python
x.shape
```

```python
model1 = tf.models.Sequential()
## add the layers
model1.add(tf.layers.Dense(16,input_dim=30,activation='relu'))  ## 2 layers (I/p + Hidden)
model1.add(tf.layers.Dense(32,activation='relu'))   ## next hidden layer
model1.add(tf.layers.Dense(1,activation='sigmoid'))  ## output
## compile the model
model1.compile(optimizer='adam',loss="binary_crossentropy",metrics=['accuracy'])
```

```python
model2 = tf.models.Sequential()
## add the layers
model2.add(tf.layers.Dense(16,input_dim=30,activation='relu'))  ## 2 layers (I/p + Hidden)
model2.add(tf.layers.Dense(32,activation='relu'))   ## next hidden layer
model2.add(tf.layers.Dense(2,activation='softmax'))  ## output
## compile the model
model2.compile(optimizer='adam',loss="sparse_categorical_crossentropy",metrics=['accuracy'])
```

```python
## Train the model
model1.fit(xtrain,ytrain,epochs=500)
```

```python
model2.fit(xtrain,ytrain,epochs=500)
```

```python
### Prediction
```

```python
ypred1 = model1.predict(xtest)
ypred1 = ypred1.round()
ypred1
```

```python
ypred2 = model2.predict(xtest)
ypred2 = np.argmax(ypred2,axis=1)
ypred2
```

```python
### Evaluate
```

```python
len(xtest)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
```

```python
cm1 = confusion_matrix(ytest,ypred1)
cm1
```

```python
cm2 = confusion_matrix(ytest,ypred2)
cm2
```

```python
ac1 = accuracy_score(ytest,ypred1)
ac1
```

```python
ac2 = accuracy_score(ytest,ypred2)
ac2
```
